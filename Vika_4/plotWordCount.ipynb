```python
import re
import numpy as np 

def add(x,D):
    # virkni: b√¶tir staki √≠ dictionary og telur fj√∂lda staka
    # inntak: strengur x og Dictionary D
    # √∫ttak: ekkert
    if x in D:
        D[x] = D[x]+1
    else:
        D[x] = 1
    return

def getWordsFromFile(fileName):
    # virkni: les skr√°, hreinsar og einangrar or√∞in
    # inntak: skr√°arheiti
    # √∫ttak: or√∞in
    
    # lesa inn skr√°na
    myFile = open(fileName, 'r')
    texti = myFile.read()
    texti = re.sub(r'[-0-9{}&+()"=!.?:/|¬ª¬©><#¬´,_+;%\[\]@$*\'\\^`~]', '', texti)
    texti = texti.lower()
    words = np.array(texti.split())
    return words

# __main__    
# Lesum inn skr√° og teljum hva√∞ hvert or√∞ kemur fyrir oft √≠ skr√°nni
myFileName = 'laxdaela_saga.txt'
words = getWordsFromFile(myFileName)
wordCount = {}
for word in words:
    add(word,wordCount)
print(wordCount)
```

```python
# Byrjum h√©r!
import operator as op
import matplotlib.pyplot as plt

sortedWordCount = sorted(wordCount.items(), key=op.itemgetter(1), reverse=True)

# Breyta √≠ or√∞ab√≥k
sortedWordCountDict = dict(sortedWordCount)
words = list(sortedWordCountDict.keys())
wordCounts = list(sortedWordCountDict.values())

# # ----- Barplot -----
# plt.bar(words[:40],wordCounts[0:40])
# plt.xticks(rotation = 90)
# plt.show()

# # ----- Horizontal Barplot -----
plt.figure(figsize=[12,10])
plt.barh(words[0:40][::-1], wordCounts[0:40][::-1])
plt.show()
```
